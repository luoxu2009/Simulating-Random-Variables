#### In this notebook, I will explore generating random variables with various distributions from Random Uniform (0,1) distributions
1. Generate Random Variable (RV) that follows exponential distribution with Inverse Transform Method
2. Generate RV that follows Beta(4,3) distribution with Acceptance-Rejection Method
3. Generate RV that follows Normal distribution with the Box-Muller Transform method
4. Generate RV that follows Poisson Distribution starting from exponential distributed RV

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 1. Generate the exponential distribution rate * exp(-rate * x), x in (0, +inf)

```python
### define the function that generate exponentially distributed random numbers. Takes 3 inputs N_exp, rate (or lambda)
### and Random Seed, return a list of the randomly generated numbers that follow the exponential distribution.

def ExpGenerator(N_exp, rate, seed=0):
    np.random.seed(seed=seed)
    Nums=[]
    for i in range(N_exp):
        U=np.random.rand()
        ### recall that using the Inverse transform method, the inverse transform of c.d.f of exponential distribution evaluated
        ### at U is the number we want to generate: X
        X=-np.log(U)/rate
        Nums.append(X)
    return Nums
```

```python
Numslist=ExpGenerator(10000, 1, seed=42)
```

```python
### create a PDF for exponential dist, will be used to compare to our generated dist.
x=np.linspace(0, 10, 1000)
y=np.exp(-x)
```

```python
plt.figure(figsize=(8,6))
plt.hist(np.array(Numslist), bins=100, density=True, label='hist of Generated Exp Dist')
plt.plot(x,y, label='P.D.F of Exp Dist')
plt.legend()
plt.xlabel('Generated X')
plt.ylabel('Probability')
plt.show()
```

#### we can see that the distribution of the random variable we generated from ExpGenerator follows the theoretical P.D.F of exponential distribution nicely!

#### 2. Generate Beta(4, 3) distribution, which has p.d.f: f(x)=60 * x^3 * (1-x)^2 for x in (0,1), using Acceptance-Rejection Method

```python
### Note peak of Beta(4,3) is at 3/5 with value 2.0736, set our major function at g(x)=2.0736 * U(x), U(x) is uniform function.
### Basically we can generate U1~U(0,1), then evaluate another independent U2~U(0,1) against f(U1)/g(U1)
### accept X=U1 only when U2 <= f(U1)/g(U1), X generated then will follow f(x) as its distribution

def BetaGenerator(N_exp, seed=0):
    np.random.seed(seed=seed)
    Nums=[]
    for i in range(N_exp):
        U1=np.random.rand()
        U2=np.random.rand()
        if U2 <= (60 * U1**3 * (1-U1)**2)/2.0736:
            Nums.append(U1)
    return Nums
```

```python
Numslist2=BetaGenerator(100000, seed=42)
```

```python
from scipy.stats import beta
```

```python
x=np.linspace(0,1,1000)
y=beta.pdf(x,4,3)
```

```python
plt.figure(figsize=(8,6))
plt.hist(np.array(Numslist2), bins=100, density=True, label='hist of Generated Beta(4,3) Dist')
plt.plot(x,y, label='P.D.F of Beta(4,3) Dist')
plt.legend()
plt.xlabel('Generated X')
plt.ylabel('Probability')
plt.show()
```

#### Looks like our A-R method generated Beta Dist matches the theoretical Beta(4,3) p.d.f pretty well!

#### 3. Generate Normal distribution with the Box-Muller Transform method.

```python
from math import pi
```

```python
#### Idea behind Box Muller Transform is that sum of squares of two Standard Normal RV follows chi square distribution
#### with 2 degrees of freedom, remember that chi-square dist with two DOF is just an exponential dist with lambda of 1/2.
#### so one can generate R^2=Z1^2+Z2^2 from -2log(U1), where U1~U(0,1), then Z1 follows from R*cos(theta), while theta is another
#### independent uniform dist with theta~2*pi*U2, where U2~U(0,1).

def SNGenerator(N_exp, mu, sigma, seed=0):
    np.random.seed(seed=seed)
    Nums=[]
    for i in range(N_exp):
        U1=np.random.rand()
        U2=np.random.rand()
        R=np.sqrt(-2*np.log(U1))
        T=2*pi*U2
        Z=R*np.cos(T)
        X=sigma*Z+mu ### remember that if Z~N(0,1), then mu+sigma*Z ~ N(mu, sigma**2)
        Nums.append(X)
    return Nums
        
    
```

```python
Numslist3=SNGenerator(50000, 5, 2, seed=42)
```

```python
from scipy.stats import norm
```

```python
x=np.linspace(-3,13,1000)
y=norm.pdf(x, 5, 2)
```

```python
plt.figure(figsize=(12,10))
plt.hist(np.array(Numslist3), bins=100, density=True, label='hist of Generated Normal(5,2^2) Dist')
plt.plot(x,y, label='P.D.F of Normal(5,2^2) Dist')
plt.legend(loc='best')
plt.xlabel('Generated X')
plt.ylabel('Probability')
plt.show()
```

#### looks like it is working well!

#### 4. Simulate Poisson Distribution

```python
### remember that the time interval between two Poisson events follows exponential dist.
### so for Poisson dist with rate lambda, we can generate time intervals that follow Exp dist t1, t2,...
### once t1+t2+...>1 (unit time interval), we return the event number as X, which follows Poisson Dist with rate lambda.
def PoissonGenerator(rate):
    event=0
    T=0
    while T<1:
        U=np.random.rand()
        t=-np.log(U)/rate
        T=T+t
        event=event+1
    return event-1

def PoissonSimulator(N_exp, rate, seed=0):
    np.random.seed(seed=seed)
    Nums=[]
    for i in range(N_exp):
        X=PoissonGenerator(rate)
        Nums.append(X)
    return Nums

```

```python
Numslist4=PoissonSimulator(10000, 5, seed=42)
```

```python
from scipy.stats import poisson
```

```python
x=np.linspace(0, 20 ,21)
y=poisson.pmf(x, 5)
```

```python
plt.figure(figsize=(12,10))
plt.hist(np.array(Numslist4), bins= range(21), density=True, align='left', label='hist of Generated Poisson(5) Dist')
plt.plot(x,y,'ro-', label='P.M.F of Possion(5) Dist')
plt.legend(loc='best')
plt.xlabel('Generated X')
plt.ylabel('Probability')
plt.show()
```

#### looks like the simulated Poisson distribution matches the p.m.f of the corresponding Poisson Dist!
